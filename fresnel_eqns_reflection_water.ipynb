# Fresnel equations: reflection coefficients

**PURPOSE:** Application of Fresnel equations for the case of reflection from air (n1=1) and
water (n2=1.33). Brewster angle.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Refraction indices
n1 = 1    # air
n  = 1.33 #Â water
```

```python
# Incident angle, gamma
ndata     = 50000 
gamma_deg = np.linspace(0.0, 90.0, num=ndata) 
gamma     = gamma_deg * np.pi/180.0
```

```python
# rpar, rper:
raiz = np.sqrt(n**2 - np.sin(gamma)**2 )
rpar = (n**2 * np.cos(gamma) - raiz ) / (n**2 * np.cos(gamma) + raiz )
rper = (       np.cos(gamma) - raiz ) / (       np.cos(gamma) + raiz )
```

```python
# Output Stokes parameters:
Ip = 0.5 * (rpar**2 + rper**2)
Qp = 0.5 * (rpar**2 - rper**2)
ratio = Qp/Ip
```

```python
# Brewster angle (numerical evaluation)
print(np.min(ratio))
di = np.where(ratio == np.min(ratio))
ang_brewster = np.min(gamma_deg[di])
print('Brewster angle (deg) = ',ang_brewster)
```

```python
# Plot
plt.plot( gamma_deg, Ip,'-b',label='I')
plt.plot( gamma_deg, Qp,'-r',label='Q')
plt.plot( gamma_deg, Qp/Ip,'-g',label='Q/I')
plt.axvline(ang_brewster,label='Brewster angle',color='yellow')
plt.legend()
plt.xlabel('Incident angle (deg)')
plt.ylabel('Stokes parameters / Ratio')
```
